```python
import random
import matplotlib.pyplot as plt

with open('without5Gcore/uploadspeed.txt', 'r') as file:
    content = file.read().splitlines()
uploadspeedwithoutcore = [float(item) for item in content]

with open('Free5gc/uploadspeed.txt', 'r') as file:
    content = file.read().splitlines()
uploadspeedfree5gc = [float(item) for item in content]

with open('Open5gs//uploadspeed.txt', 'r') as file:
    content = file.read().splitlines()
uploadspeedopen5gs = [float(item) for item in content]
```

```python
plt.figure()
plt.plot([i for i in range(15)], uploadspeedwithoutcore, label='without core network')
plt.plot([i for i in range(15)], uploadspeedfree5gc, label='Free5gc')
plt.plot([i for i in range(15)], uploadspeedopen5gs, label='Open5gs')

plt.xlabel('Time (s)')
plt.ylabel('Uplink (Mbps)')
plt.title('Uplink Throughput')
plt.ylim(5, 35)

plt.legend()
plt.show()
```

```python
with open("Free5gc/cpu.txt", "r") as file:
    content = file.read().splitlines()
cpufree5gc = [float(item) for item in content]

with open("Open5gs//cpu.txt", "r") as file:
    content = file.read().splitlines()
cpuopen5gs = [float(item) for item in content]
```

```python
plt.figure()
plt.plot(cpufree5gc, label="Free5gc")
plt.plot(cpuopen5gs, label="Open5gs")
# plt.plot([i for i in range(15)], latencywithcorenetwork, label='with core network')

plt.xlabel('Time (s)')
plt.ylabel('CPU (percentage)')
plt.ylim(0, 90)
plt.title('CPU Consumption')

plt.legend()
plt.show()
```

```python
under_load_cpu_consumption_free5gc = [i for i in cpufree5gc if i > 10]
mean_cpu_consumption_free5gc = sum(under_load_cpu_consumption_free5gc) / len(under_load_cpu_consumption_free5gc)
print("Mean CPU Consumption for Free5gc: ", mean_cpu_consumption_free5gc)

under_load_cpu_consumption_open5gs = [i for i in cpuopen5gs if i > 10]
mean_cpu_consumption_open5gs = sum(under_load_cpu_consumption_open5gs) / len(under_load_cpu_consumption_open5gs)
print("Mean CPU Consumption for Open5gs: ", mean_cpu_consumption_open5gs)
```
