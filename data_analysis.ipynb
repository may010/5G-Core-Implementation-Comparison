```python
import random
import matplotlib.pyplot as plt

with open('performance_metrics/uploadspeedwithoutcore.txt', 'r') as file:
    content = file.read().splitlines()
uploadspeedwithoutcore = [float(item) for item in content]

with open('performance_metrics/uploadspeedwithcore.txt', 'r') as file:
    content = file.read().splitlines()
uploadspeedwithcore = [float(item) for item in content]
```

```python
plt.figure()
plt.plot([i for i in range(15)], uploadspeedwithoutcore, label='without core network')
plt.plot([i for i in range(15)], uploadspeedwithcore, label='with core network')

# Add labels to the axes and a title
plt.xlabel('Time (s)')
plt.ylabel('Uplink (Mbps)')
plt.title('Uplink Throughput')
plt.ylim(15, 35)

plt.legend()
plt.show()
```
